# Activation Functions

*   Applied to the output of a neuron (or layer of neurons)
*   If the activation function itself is nonlinear, it allows for neural networks with usually two or more hidden layers to map nonlinear functions.


*   Two types of activation functions:
    *   activation function used in hidden layers (often the same for layers)
    *   activation function used in output layer

## The Step Activation Function

```python
x = np.linspace(-2, 2, 500)
y = [1 if i>=0 else 0 for i in x]
```

```python
x_axis = [0]*len(y)
y_axis = [0]*len(x)
```

```python
plt.plot(x_axis,x)
plt.plot(x,y_axis)
plt.plot(x, y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('The Step Activation FUnction')
plt.show()
```

## The Sigmoid Activation Function

*   Its hard to tell how close the step function is to activating or deactivating. Maybe it is very close, or maybe it is very far.


*   In sigmoid activation function, were getting a value that can be
reversed to its original value; the returned value contains all the information from the input.

* The output from the Sigmoid function, being in the range of 0 to 1, also works
better with neural networks  especially compared to the range of the negative to the positive infinity  and adds nonlinearity.



```python
x = np.linspace(-3, 3, 10000)
y = [1 / (1 + np.exp(-i)) for i in x]
```

```python
x_axis = [0]*len(y)
y_axis = [0]*len(x)
```

```python
plt.plot(x_axis,x)
plt.plot(x,y_axis)
plt.plot(x, y)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.title('The Sigmoid Activation Function')
plt.grid()
plt.show()
```

## Calculating Network Error with Loss



*   The **loss function**, also referred to as the **cost function**, is the algorithm that quantifies how wrong a model is.

*   **Loss** is the measure of this metric. Since loss is the models error, we ideally want it to be 0.



### Categorical Cross-Entropy Loss

*   Categorical cross-entropy is explicitly used to compare a ground-truth probability (y o r  targets ) and some predicted distribution (y-hat o r predictions).

*   It is also one of the most commonly used loss functions with a softmax activation on the output layer.



Categorical Cross Entropy Loss formula:

$$
L_i = - \sum_{j} y_{i,j} \cdot \log(\hat{y}_{i,j})
$$

Where:
- $L_i$ denotes sample loss value, *i* is the *i-th* sample in the set, *j* is the label/output index, *y* denotes the target values, and *y-hat* denotes the predicted values.

```python
output_probs = torch.tensor([0.4, 0.9, 0.1])
target_values = torch.tensor([0, 1, 1])
```

```python
sum_values = 0
for i,j in zip(output_probs,target_values):
  sum_values += j*torch.log(i)
```
